```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('autos.csv')
```

# 1. Grafik Batang

```python
plt.figure(figsize=(14, 6))
sns.countplot(data=data, x='brand', palette='magma', order=data['brand'].value_counts().index)
plt.title('Distribusi Merek Mobil')
plt.xlabel('Merek Mobil')
plt.ylabel('Jumlah Mobil')
plt.xticks(rotation=90)
plt.show()
```

Volkswagen adalah merek mobil yang paling umum dalam dataset ini, diikuti oleh BMW, Opel, dan Mercedes-Benz.

# 2. Pie Chart

```python
# Menghitung jumlah data untuk setiap kategori dalam kolom 'abtest'
abtest_counts = data['abtest'].value_counts()

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(abtest_counts, labels=abtest_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribusi Data untuk Kolom abtest')
plt.xlabel('Kategori abtest')
plt.ylabel('Jumlah Data')

plt.show()
```

Dari visualisasi pie chart di atas, dapat disimpulkan bahwa data pada kolom 'abtest' terbagi secara cukup merata antara kategori 'test' dan 'control', dengan perbandingan sekitar 50:50. 

# 3. Heatmap

```python
# Membuat visualisasi dengan heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Korelasi antara Variabel dalam Dataset Mobil Bekas')
plt.xlabel('Variabel X')
plt.ylabel('Variabel Y')
plt.show()
```

Visualisasi heatmap ini membantu kita memahami sejauh mana variabel-variabel dalam dataset ini berkorelasi satu sama lain.

# 4. Line Plot

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='yearOfRegistration', y='price')
plt.title('Harga Jual Terhadap Tahun Pembuatan Mobil')
plt.xlabel('Tahun Pembuatan')
plt.ylabel('Harga Jual')
plt.show()
```

Dari grafik ini, kita dapat melihat tren umum bahwa harga jual mobil cenderung meningkat seiring dengan tahun pembuatan. Ini menunjukkan bahwa mobil yang lebih baru umumnya memiliki harga jual yang lebih tinggi.
